```python
from PIL import Image
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np

A = imread('images/sf-gray.jpg')
```

```python
img = plt.imshow(A)
img.set_cmap('gray')
plt.show()
```

```python
U, S, VT = np.linalg.svd(A,full_matrices=False)
S = np.diag(S)

j = 1
for k in (50, 70, 90, 100, 120, 150):
    # appr image
    X = U[:,:k] @ S[0:k,:k] @ VT[:k,:]
    plt.figure(j)
    img = plt.imshow(X)
    img.set_cmap('gray')
    plt.title('k = ' + str(k))
    plt.show()
    j += 1
```

```python
# Best Approximation ~k = 90
k = 90
B = U[:,:k] @ S[0:k,:k] @ VT[:k,:]
img  = plt.imshow(B)
img.set_cmap('gray')
plt.axis('off')

plt.savefig('images/sf-compressed.jpg')
```
